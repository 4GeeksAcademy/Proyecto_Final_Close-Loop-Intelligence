# Definicion del Problema
---

Muchas empresas sufren de dos males: **el stockout** (perder ventas por no tener producto) o **el exceso de inventario** (dinero muerto en estanterías que se deprecia). Este proyecto construye un Sistema de Inteligencia de Ciclo Cerrado que predice cuánto se va a vender para decidir exactamente cuánto comprar y qué promocionar para salir del inventario lento.

La propuesta, consiste en crear un modelo de datos almacenado en SQL que contenga informacion basica de cualquier empresa dedicada a la compra-venta de bienes materiales:

1. Tabla de Ventas.
    Debe contener columnas como: ID_Transaccion, Fecha, ID_Producto, Cantidad, Precio_Unitario, ID_Cliente.

2. Tabla de Productos.
    Debe contener columnas como: ID_Producto, Categoria, Costo_Reposicion, Stock_Actual, Punto_Reorden. 

A partir de esta informacion, se propone crear dos modelos para manejar el siguiente enfoque: 

1. Enfoque de Ventas (Prediccion o Forecasting): 

    Inicialmente, se pretende emplear Series de Tiempo y/o un modelo de Regresion (como Random Forest) que sea capaz de predecir la cantidad de unidades que se venderan de un determinado producto en los proximos 15-30 dias considerando estacionalidad y tendencias historicas.

    **Variable Objetivo:** Cantidad de Productos en Ordenes de Ventas

    **Posibles Variables Predictoras:** Fecha, Cliente, Vendedor, Precio, Categoria, Linea, Reseñas del Producto, Localizacion del Producto, Tiempos de Envio, Imagen, Precio de Envio, Peso del Producto, Volumen del Producto, Ciudad del Cliente, Ciudad del Vendedor

    A partir de la prediccion de la cantidad de productos se puede estimar a su vez una proyeccion de ventas considerando estacionario el precio de cada producto durante el marco temporal de la prediccion. 

2. Enfoque de Compras/Inventario (Optimizacion):

    Considerando las cantidades vendidas de un producto dado en un marco temporal especifico, se clasificaran los productos de mayor a menor rotacion de inventario. Para ello, se puede emplear la tecnica de Naive-Bayes dado que la rotacion de productos puede variar en su clasificacion dependiendo del periodo en el que se este evaluando (estacionalidad).

    El stock minimo necesario, puede ser calculado de forma estadistica a partir de la rotacion del producto en su data de venta historica.

    $$Stock\ Mínimo = (Demanda\ Promedio \times Tiempo\ de\ Entrega) + Stock\ de\ Seguridad$$

    $$Stock\ de\ Seguridad = Z \times \sigma \times \sqrt{L}$$

    **Variable Objetivo:** Categoria de Rotacion

    **Posibles Variables Predictoras:** Mes, Semana, Dia, Precio del Producto, Categoria del Producto, Reseña del Producto, Tiempos de Entrega, Cliente, Ciudad del Cliente, Vendedor, Ciudad del Vendedor, Precio de Envio, Imagen, Peso del Producto, Volumen del Producto.

Se consolidara todo el conjunto de datos en un DataFrame maestro compuesto de: Fecha, ID Orden, ID Cliente, Nombre del Cliente, Ciudad del Cliente, Vendedor, Ciudad del Vendedor, ID Producto, Categoria del Producto, Peso (g), Volumen (cm3), Fotos del Producto, Stock Minimo, Precio Unitario, Cantidad Ordenada, Subtotal de la Orden, Costo del Flete, Total de la Orden, Total del Pago, Estado de la Orden, Calificacion de la Orden.
